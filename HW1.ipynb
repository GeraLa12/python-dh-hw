Для первого слова, которое должно читаться одинаково, выведите YES, а для второго слова, которое должно читаться неодинаково, -- NO.

```python
a = input ("Напишите слово: ")
b = a [ : :-1]

if a == b:
    print ("Yes")
else:
    print ("No")
```
